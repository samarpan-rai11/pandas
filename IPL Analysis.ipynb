# Function for generating matches list in a city (VVI Function)

```python
def city_match_count(city):     #VVI function 
    mask = data['city'] == city
    return data[mask]
```

```python
city_match_count('Pune')
```

```python
city_match_count('Pune').shape
```

```python
city_match_count('Rajkot')      #all matches that held in 'Rajkot' city
```

```python
mask1 = data['city'] == 'Rajkot'
mask2 = data['season'] > 2016

data[mask1 & mask2]      #in boolean value we use '&' bitwise operator
```

# sort_values( ) method

```python
#sorting above value in descending order by making ascending value false

(data['team1'].value_counts() + data['team2'].value_counts()).sort_values(ascending=False)
```

```python
#sorting dataset according to city and it automatically sorts it alphabetically 
#we sorted the below data opposite of alphabetically

data.sort_values('city',ascending=False).head()
```

```python
#sorting dataset from two different columns 'city' & 'date'
#'inplace' parameter is used to change the dataset permanently

data.sort_values(['city','date'],ascending=[True,False]).head(10)
```

# drop_duplicates( ) method

This deletes all duplicates mentioned in subset parameter but only temporarily

```python
#'inplace=True' parameter is used to change the dataset permanently

data.drop_duplicates(subset=['city']).shape
```

```python
data.drop_duplicates(subset=['city','season']).shape
```

```python
#we don't need to use subset for only one series
#the below function will give us last match of every season

data.drop_duplicates('season',keep='last')
```

```python
#winner of every season of IPL

data.drop_duplicates('season',keep='last')[['season','winner']].sort_values('season')  #sorting by season too
```

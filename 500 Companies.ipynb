# The groupby( ) function

groupby( ) function creates group of the given series like below of 'Sector'. It is for categorical data that repeats itself multiple time and this function helps to group them

```python
sectors = company.groupby('Sector')
```

```python
len(sectors)
```

```python
#size function gives total no. of each sector that we grouped

sectors.size().sort_values(ascending=False)
```

```python
sectors.first()   #this function will give first company of each sector 
```

```python
sectors.last()     #this function will give last company of each sector 
```

```python
#this will give us dictionaries of sectors where key is sector and value is index of the companies

sectors.groups 
```

```python
#this function will give every company of the sector passed in function

sectors.get_group('Apparel')
```

```python
#sum of profits of every sectors

sectors['Profits'].sum().sort_values(ascending=False)
```

```python
#mean of profits of every sectors

sectors['Profits'].mean().sort_values(ascending=False)
```

```python
sectors[['Revenues','Profits']].sum()
```
